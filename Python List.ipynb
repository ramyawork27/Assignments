### Question 10:  Write a Python program to sort a list of strings in Ascii Order [Ascending order]


- `sample input : `
    - `var = ['a' , 'c' , 'b']`
- `sample output :`
    - `result : ['a' , 'b' , 'c']`


```python
n= int(input("maxmim numbers to store in list1 \t"))
a = []
for i in range(0,n):
  a.append(input("enter a character \t"))

for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if ord(a[i]) > ord(a[j]):
            temp = a[i]
            a[i] = a[j]
            a[j] = temp

print(f"sorted order of the list {a}")
```
